```python
input_folder = "D:/mine"
output_folder = "D:/cv_final_project/mine_grayscale"
os.makedirs(output_folder, exist_ok=True)

def convert_and_save_grayscale(input_folder, output_folder):
    for root, dirs, files in os.walk(input_folder):
        for filename in files:
            # Check if the file is an image (you can add more image formats if needed)
            if filename.endswith((".jpg", ".jpeg", ".png", ".bmp")):
                # Construct the input and output paths
                input_path = os.path.join(root, filename)
                rel_path = os.path.relpath(input_path, input_folder)
                output_path = os.path.join(output_folder, rel_path)
                output_dir = os.path.dirname(output_path)
                
                os.makedirs(output_dir, exist_ok=True)
                image = cv2.imread(input_path)
                grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                cv2.imwrite(output_path, grayscale_image)

convert_and_save_grayscale(os.path.join(input_folder, "train"), os.path.join(output_folder, "train"))
convert_and_save_grayscale(os.path.join(input_folder, "test"), os.path.join(output_folder, "test"))
```

```python
input_folder = "D:/cv_final_project/mine_grayscale"
output_folder = "D:/cv_final_project/mine_grayscale_resized"
os.makedirs(output_folder, exist_ok=True)

target_width = 128
target_height = 128

def resize_and_save_grayscale_images(input_folder, output_folder, target_width, target_height):
    for root, dirs, files in os.walk(input_folder):
        for filename in files:
            # Check if the file is an image (you can add more image formats if needed)
            if filename.endswith((".jpg", ".jpeg", ".png", ".bmp")):
                # Construct the input and output paths
                input_path = os.path.join(root, filename)
                rel_path = os.path.relpath(input_path, input_folder)
                output_path = os.path.join(output_folder, rel_path)
                output_dir = os.path.dirname(output_path)

                os.makedirs(output_dir, exist_ok=True)
                grayscale_image = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
                
                # Resize the grayscale image to the target dimensions
                resized_image = cv2.resize(grayscale_image, (target_width, target_height))

                # Save the resized grayscale image
                cv2.imwrite(output_path, resized_image)

resize_and_save_grayscale_images(input_folder, output_folder, target_width, target_height)
```

```python
train_grayscale_folder = "D:/cv_final_project/mine_grayscale_resized/train_grayscale_resized"
test_grayscale_folder = "D:/cv_final_project/mine_grayscale_resized/test_grayscale_resized"

kernel_size = (5, 5)

def apply_blurring(input_folder, output_folder):
    for animal_folder in os.listdir(input_folder):
        animal_path = os.path.join(input_folder, animal_folder)
        output_animal_path = os.path.join(output_folder, animal_folder)
        os.makedirs(output_animal_path, exist_ok=True)

        for image_name in os.listdir(animal_path):
            image_path = os.path.join(animal_path, image_name)
            output_path = os.path.join(output_animal_path, image_name)

            if os.path.isfile(image_path):
                image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                blurred_image = cv2.GaussianBlur(image, kernel_size, 0)
                cv2.imwrite(output_path, blurred_image)

apply_blurring(train_grayscale_folder, "D:/cv_final_project/train_blurred")
apply_blurring(test_grayscale_folder, "D:/cv_final_project/test_blurred")
```

```python
lower_threshold = 50
upper_threshold = 150
```

```python
train_blurred_folder = "D:/cv_final_project/train_blurred"
test_blurred_folder = "D:/cv_final_project/test_blurred"

def apply_canny_edge_detection(input_folder, output_folder):
    for animal_folder in os.listdir(input_folder):
        animal_path = os.path.join(input_folder, animal_folder)
        output_animal_path = os.path.join(output_folder, animal_folder)
        os.makedirs(output_animal_path, exist_ok=True)

        for image_name in os.listdir(animal_path):
            image_path = os.path.join(animal_path, image_name)
            output_path = os.path.join(output_animal_path, image_name)

            if os.path.isfile(image_path):
                image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                edges = cv2.Canny(image, lower_threshold, upper_threshold)
                cv2.imwrite(output_path, edges)

apply_canny_edge_detection(train_blurred_folder, "D:/cv_final_project/train_edges")
apply_canny_edge_detection(test_blurred_folder, "D:/cv_final_project/test_edges")
```

```python
train_edges_folder = "D:/cv_final_project/train_edges"
test_edges_folder = "D:/cv_final_project/test_edges"

sift = cv2.SIFT_create()

def save_sift_feature_images(input_folder, output_folder):
    for animal_folder in os.listdir(input_folder):
        animal_path = os.path.join(input_folder, animal_folder)
        output_animal_path = os.path.join(output_folder, animal_folder)
        os.makedirs(output_animal_path, exist_ok=True)

        for image_name in os.listdir(animal_path):
            image_path = os.path.join(animal_path, image_name)
            output_path = os.path.join(output_animal_path, image_name)

            if os.path.isfile(image_path):
                image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                keypoints, _ = sift.detectAndCompute(image, None)

                image_with_keypoints = cv2.drawKeypoints(image, keypoints, None)

                cv2.imwrite(output_path, image_with_keypoints)

save_sift_feature_images(train_edges_folder, "D:/cv_final_project/train_sift_feature_images")
save_sift_feature_images(test_edges_folder, "D:/cv_final_project/test_sift_feature_images")
```

```python
train_edges_folder = "D:/cv_final_project/train_edges"
test_edges_folder = "D:/cv_final_project/test_edges"

import numpy as np
sift = cv2.SIFT_create()
def save_sift_descriptors(input_folder, output_folder):
    for animal_folder in os.listdir(input_folder):
        animal_path = os.path.join(input_folder, animal_folder)
        output_animal_path = os.path.join(output_folder, animal_folder)
        os.makedirs(output_animal_path, exist_ok=True)

        for image_name in os.listdir(animal_path):
            image_path = os.path.join(animal_path, image_name)
            output_path = os.path.join(output_animal_path, image_name)

            if os.path.isfile(image_path):
                image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                keypoints, descriptors = sift.detectAndCompute(image, None)

                if descriptors is not None:
                    output_descriptor_path = os.path.splitext(output_path)[0] + ".txt"
                    np.savetxt(output_descriptor_path, descriptors, delimiter=' ', fmt='%f')

save_sift_descriptors(train_edges_folder, "D:/cv_final_project/train_sift_descriptors")
save_sift_descriptors(test_edges_folder, "D:/cv_final_project/test_sift_descriptors")
```

```python
import cv2
import os
import pandas as pd

input_folder = "D:/cv_final_project/train_sift_feature_images"  # Update this to your SIFT feature images folder
output_csv_path = "D:/cv_final_project/sift_descriptors.csv"  # Update this to your desired output CSV file path

sift = cv2.SIFT_create()

# Initialize an empty DataFrame to store the descriptors
df_descriptors = pd.DataFrame()

for animal_folder in os.listdir(input_folder):
    animal_path = os.path.join(input_folder, animal_folder)

    for image_name in os.listdir(animal_path):
        image_path = os.path.join(animal_path, image_name)

        if os.path.isfile(image_path):
            image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            keypoints, descriptors = sift.detectAndCompute(image, None)

            if descriptors is not None:
                # Print the descriptor shape
                print(f"Descriptor shape for {image_name}: {descriptors.shape}")

                # Convert the descriptors array to a DataFrame
                df_image_descriptors = pd.DataFrame(descriptors)

                # Append the DataFrame to the main DataFrame
                df_descriptors = df_descriptors.append(df_image_descriptors, ignore_index=True)

# Save the concatenated descriptors to a CSV file
df_descriptors.to_csv(output_csv_path, index=False)
```
