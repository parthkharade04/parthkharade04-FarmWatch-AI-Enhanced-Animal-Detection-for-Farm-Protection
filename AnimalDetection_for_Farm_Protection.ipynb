```python
import numpy as np
import pandas as pd                               
import os
import csv
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn import svm, datasets
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics as metrics 
from sklearn.metrics import accuracy_score
```

```python
folder1 = "D:/mine/train/Bull"

# Initialize a counter for printing descriptor shape
i = 0

# Create an empty list to store image descriptors
data = []

# Resize dimensions
resize = (280, 430)

# Initialize SIFT detector
sift = cv2.SIFT_create()

for filename in os.listdir(folder1):
    # Path to the image
    path = os.path.join(folder1, filename)
    
    # Read the image
    img = cv2.imread(path)
    
    # Check if the image is loaded successfully
    if img is not None:
        # Resize the image
        img = cv2.resize(img, resize)
        
        # Convert the image to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Detect SIFT keypoints and descriptors
        keypoints, descriptors = sift.detectAndCompute(gray, None)
        
        # Convert the descriptor array into a DataFrame format
        out = pd.DataFrame(descriptors)
        
        # Append the descriptor data to the list
        data.append(out)
        
        # Print the descriptor shape
        print("Descriptor shape", i, ":", out.shape)
        
        i += 1

# Combine all the descriptor data into a single DataFrame
descriptor_data = pd.concat(data, ignore_index=True)

# Save the descriptor data to a CSV file
descriptor_data.to_csv("D:/cv_final_project/z/Bull.csv", index=False)

```

```python
#performing kmeans prediction of the entire bottle dataset with the pretrained kmeans model

#initialising i=0; as its the first class
i=0
data=[]
#k=0

for filename in os.listdir(folder1):
    #path 
    path=os.path.join(folder1,filename)
    a=cv2.imread(path)
    
    #resize image
    resize=(280,430)
    img=cv2.resize(a,resize)
    
    #gray image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    sift = cv2.SIFT_create()
    
    keypoints, descriptors = sift.detectAndCompute(gray, None)
    out=pd.DataFrame(descriptors)
    
    #predict values of feature vector with pretrained kmeans
    #ValueError: Buffer dtype mismatch, expected 'float', in order to avoid this dtype=np.double
    
    array_double = np.array(out, dtype=np.double)
    a=kmeans_data1.predict(array_double)
    
    hist_data1=np.histogram(a,bins=[0,1,2,3,4,5])
    
    #append the dataframe into the array in append mode, the array will only have 5 values which will store the values in a row
    data.append(hist_data1[0])
    #k=k+1
    
#convert Array to Dataframe and append to the list
Output1 = pd.DataFrame(data)
#add row class 
Output1["Class"] = 1 
csv_data=Output1.to_csv(r'D:/cv_final_project\z\Dataset1.csv', mode='a',header=False,index=False)

print("Complete")
```
